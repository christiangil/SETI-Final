```python
#working directly from breakthrough/GBT/voyager/voyager.ipynb
#Using Python 2 and the Filterbank.py method of importing filterbank files
#https://github.com/UCBerkeleySETI/breakthrough/blob/master/GBT/voyager/voyager.ipynb

import pylab as plt
from blimpy2 import Filterbank
%matplotlib inline

obs = Filterbank('voyager_f1032192_t300_v2.fil')
obs.info()
```

```python
print obs.header
print obs.data.shape
```

```python
#plot_spectrum works here.

obs.plot_spectrum()
#fig = plt.gcf()
#fig.savefig("justwork5.png", bbox_inches="tight")
```

```python
#plot_spectrum can't zoom (breaks in Filterbank case)

obs.plot_spectrum(f_start=8420.18, f_stop=8420.25)
```

```python
#plot_spectrum can't zoom (breaks in Filterbank case)

plt.figure(figsize=(8, 6))
plt.subplot(3,1,1)
obs.plot_spectrum(f_start=8420.193, f_stop=8420.195) # left sideband
plt.subplot(3,1,2)
obs.plot_spectrum(f_start=8420.2163, f_stop=8420.2166) # carrier
plt.subplot(3,1,3)
obs.plot_spectrum(f_start=8420.238, f_stop=8420.24) # right sideband
plt.tight_layout()
```

```python
#plot_waterfall works

plt.figure(figsize=(10,6))
obs.plot_waterfall() 
```

```python
#plot_waterfall can't zoom. x-axis is updated, but plot remains the same in Filterbank case

plt.figure(figsize=(8, 6))
plt.subplot(3,1,1)
obs.plot_waterfall(f_start=8420.193, f_stop=8420.195) # left sideband
plt.subplot(3,1,2)
obs.plot_waterfall(f_start=8420.2163, f_stop=8420.2166) # carrier
plt.subplot(3,1,3)
obs.plot_waterfall(f_start=8420.238, f_stop=8420.24) # right sideband
plt.tight_layout()
```
